```python
df_sample=df_resample.dropna()
df_sample.head(40)
```

```python
# plot the trend
plt.plot(df_sample.index,df_sample.high,marker='o')
plt.show()
```

```python
# plot the volume
plt.plot(df_sample.index,df_sample.volume,marker='*')
plt.show()
```

```python
# organize data - average stock high
df_stock_value=df_sample.drop(['open','low','close','volume'],axis=1)
df_stock_value.columns=['average high price']

#decompose time series
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(df_stock_value,period=5)

df_stock_value_trend = decomposition.trend.dropna()
df_stock_value_seasonal = decomposition.seasonal.dropna()
df_stock_value_residual = decomposition.resid.dropna()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
fig.tight_layout(pad=3)

ax1.plot(df_stock_value, label = 'Original',marker='o')
ax1.legend(['original'],prop={'size': 10})

ax2.plot(df_stock_value_trend, label = 'Trend',marker="*")
ax2.legend(['Trend'])

fig, (ax3, ax4) = plt.subplots(1, 2, figsize=(10, 5))
fig.tight_layout(pad=3)

ax3.plot(df_stock_value_seasonal, label = 'Seasonality')
ax3.legend(['Season'])

ax4.plot(df_stock_value_residual, label = 'Residuals')
ax2.legend(['Residue'])
```

```python
# organize data - average stock volume
df_stock_volume=df_sample.drop(['open','high','low','close'],axis=1)
df_stock_volume.columns=['average daily volume']

#decompose time series
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(df_stock_volume,period=5)

df_stock_volume_trend = decomposition.trend.dropna()
df_stock_volume_seasonal = decomposition.seasonal.dropna()
df_stock_volume_residual = decomposition.resid.dropna()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
fig.tight_layout(pad=3)

ax1.plot(df_stock_volume, label = 'Original',marker='o')
ax1.legend(['original'],prop={'size': 10})

ax2.plot(df_stock_volume_trend, label = 'Trend',marker='*')
ax2.legend(['Trend'])

fig, (ax3, ax4) = plt.subplots(1, 2, figsize=(10, 5))
fig.tight_layout(pad=3)

ax3.plot(df_stock_volume_seasonal, label = 'Seasonality')
ax3.legend(['Season'])

ax4.plot(df_stock_volume_residual, label = 'Residuals')
ax2.legend(['Residue'])
```

```python
# Import augmented dicky-fuller test function
from statsmodels.tsa.stattools import adfuller

# Run test
result = adfuller(df_stock_value_trend)

# Print test statistic
print(result[0])

# Print p-value
print(result[1])

# Print critical values
print(result[4]) 

# p> 0.05, non stationary data
```

```python
y1=df_stock_value_trend
y1_stationary=y1.diff().dropna()

# Run test
result = adfuller(y1_stationary)

# Print test statistic
print(result[0])

# Print p-value
print(result[1])

# Print critical values
print(result[4]) 

plt.plot(y1_stationary)
```

```python
# Split Training and Test Data. The training data will be before July. The test data will be those after July
# Find the last day to take the car
y1_stationary.index.max()
```

```python
# Create empty list to store search results
from statsmodels.tsa.statespace.sarimax import SARIMAX

order_aic_bic=[]

# Loop over p values from 0-10
for p in range(10):
  # Loop over q values from 0-10
    for q in range(10):
        #loop over d values from 0-10
        for d in range(10):
            # create and fit ARMA(p,q) model
            model = SARIMAX(y1_stationary, order=(p,d,q))
            results = model.fit()
        
            # Append order and results tuple
            order_aic_bic.append((p,d,q,results.aic,results.bic))
```

```python
# Construct DataFrame from order_aic_bic
order_df = pd.DataFrame(order_aic_bic, 
                        columns=['p','d','q','AIC','BIC'])

# Print order_df in order of increasing AIC
print(order_df.sort_values('AIC'))

# Print order_df in order of increasing BIC
print(order_df.sort_values('BIC'))
```

```python
model1 = SARIMAX(y1_stationary, order=(1,0,4))
results1 = model1.fit()

# Print summary
print(results1.summary())
```
